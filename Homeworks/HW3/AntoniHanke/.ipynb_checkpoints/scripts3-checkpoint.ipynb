Comments:
1. Many of the variables used in the LIME explanations overlap, with a similar importance.
2. The variables are sometimes used in 2 variants (positive frequency and zero frequency) - with inverse effects.
2. The data in our dataset is sparse - the frequency of most words is 0, but it is larger for some observations. This is represented in the LIME explanations, as for example only Observation3 has a positive word frequency for *remove*, and hence only here in LIME it shows *word_freq_remove>0.00* and has a positive impact on the prediction.
3. Keeping the above points in mind, for all 4 observations the same variables are used in the explanation. The only exception is *capital_run_length_average>3.71* used only for Observation1 in place of the *word_freq_meeting* variable.

# Appendix
## Data preparation

```python
import numpy as np
import pandas as pd
import dalex as dx
import lime

spambase = pd.read_csv("spambase.csv")
```

```python
df = spambase.drop(spambase.columns[0], axis=1) #Cleaning first column which is just index
```

```python
df.describe()
```

```python
X = df.loc[:, df.columns != 'TARGET']
```

```python
y = df.loc[:, df.columns == 'TARGET']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=2)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
kf = KFold(n_splits = 5)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF_final = RandomForestClassifier(n_estimators=200, max_depth = 8, max_features = 0.3, random_state = 1).fit(X_train, y_train)
print("Train accuracy: ", accuracy_score(y_train, RF_final.predict(X_train)))
```

```python
RFexplainer = dx.Explainer(RF_final, X_train, y_train)
```

```python
RFexplainer.model_performance()
```

```python
RF_observations = []
RF_predictions = []
RF_explanation_results = []

for i in range(4):
    observation = X_train.iloc[[i]]
    RF_observations.append(observation)
    RF_predictions.append(RFexplainer.predict(observation))
    explanation = RFexplainer.predict_surrogate(observation)
    RF_explanation_results.append(explanation.result)
    explanation.plot()
```

```python
RF_predictions
```

```python
RF_explanation_results[0]
```

```python
from functools import reduce
def merge_dataframes(left, right):
    return pd.merge(left, right, on='variable', how='outer')

merged_df = reduce(merge_dataframes, RF_explanation_results)

merged_df.columns = ['variable']+['effect_obs' + str(i) for i in range(len(RF_explanation_results))]

#merged_df.fillna(0, inplace=True)
merged_df = merged_df.set_index(['variable'])
print(merged_df)
```

```python
import matplotlib.pyplot as plt
merged_df.plot(kind='barh', figsize=(10, 6))
plt.xlabel('Effect')
plt.ylabel('Variables')
plt.title('Bar Plot for effects of variables on different observations')
plt.xticks(rotation=45)
plt.gca().invert_yaxis()
plt.show()
```

## TabPFN

```python
from tabpfn import TabPFNClassifier
```

```python
tabpfn_final = TabPFNClassifier(device='cpu', N_ensemble_configurations=32).fit(X_train[:1000], y_train[:1000])
print("Train accuracy: ", accuracy_score(y_train[:1000], tabpfn_final.predict(X_train[:1000])))
```

```python
TabPFNexplainer = dx.Explainer(tabpfn_final, X_train[:1000], y_train[:1000])
```

```python
TabPFNexplainer.model_performance()
```

```python
TabPFN_observations = []
TabPFN_predictions = []
TabPFN_explanation_results = []

for i in range(4):
    observation = X_train.iloc[[i]]
    TabPFN_observations.append(observation)
    TabPFN_predictions.append(TabPFNexplainer.predict(observation))
    explanation = TabPFNexplainer.predict_surrogate(observation)
    TabPFN_explanation_results.append(explanation.result)
    explanation.plot()
```

```python
TabPFN_merged_df = reduce(merge_dataframes, TabPFN_explanation_results)

TabPFN_merged_df.columns = ['variable']+['effect_obs' + str(i) for i in range(len(TabPFN_explanation_results))]

#merged_df.fillna(0, inplace=True)
TabPFN_merged_df = TabPFN_merged_df.set_index(['variable'])
print(TabPFN_merged_df)
```

```python
TabPFN_merged_df.plot(kind='barh', figsize=(10, 6))
plt.xlabel('Effect')
plt.ylabel('Variables')
plt.title('Bar Plot for effects of variables on different observations - TabPFN')
plt.xticks(rotation=45)
plt.gca().invert_yaxis()
plt.show()
```
